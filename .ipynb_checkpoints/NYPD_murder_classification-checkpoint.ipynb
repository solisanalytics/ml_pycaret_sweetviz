# NYPD Shooting Incident - Murder Classification Using PyCaret

### Importing Libraries

```python
import pandas as pd
import numpy as np
import plotly.express as px
import sweetviz as sv
import time

from pycaret.classification import ClassificationExperiment
from sklearn.utils import resample
```

### Exploring Dataset

```python
# reading csv file
pd.set_option('display.max_columns', None)
df_nypd = pd.read_csv('NYPD_shooting_incident_data_historic.csv')
df_nypd.head()
```

```python
# displaying number of records and columns
print(df_nypd.shape)
```

```python
# displaying data types
df_nypd.dtypes
```

```python
# count of perpetuator race
df_nypd['PERP_RACE'].value_counts()
```

```python
# filter based on perpetuator race
query = "PERP_RACE in ('BLACK', 'WHITE HISPANIC', 'UNKNOWN', 'BLACK HISPANIC', 'WHITE', 'ASIAN / PACIFIC ISLANDER')"
df_nypd = df_nypd.query(query)

df_nypd['PERP_RACE'].value_counts()
```

```python
# creating new perpetrator identifier to be used as target
df_nypd['same_race_murder'] = np.where(df_nypd['PERP_RACE'] == df_nypd['VIC_RACE'], 1, 0)
df_nypd['same_race_murder'].value_counts()
```

```python
# dropping columns
cols_drop = [
    'INCIDENT_KEY', 'X_COORD_CD', 'Y_COORD_CD', 
    'Latitude', 'Longitude', 'Lon_Lat',
    'PERP_RACE', 'OCCUR_DATE', 'LOC_OF_OCCUR_DESC',
    'PRECINCT', 'OCCUR_TIME']

df_nypd = df_nypd.drop(cols_drop, axis=1)
print(df_nypd.columns)
```

```python
# performing EDA
feature_config = sv.FeatureConfig(force_num=['same_race_murder'])
analyze_report = sv.analyze([df_nypd, 'NYPD_muder'], target_feat='same_race_murder', feat_cfg=feature_config)
analyze_report.show_html(filepath='NYPD_murder_report.html')
```

### Model Training and Evaluation

```python
# creating classification object
s = ClassificationExperiment()
s.setup(
    df_nypd, 
    target='same_race_murder', 
    session_id=123,
    fold_strategy='stratifiedkfold',
    fold=10,
    fold_shuffle=True,
    numeric_imputation='median',
    categorical_imputation='model',
    remove_outliers=True,
    outliers_threshold=0.05,
    normalize=True,
    normalize_method='robust',
    remove_multicollinearity=True,
    multicollinearity_threshold=0.90)
```

```python
# comparing models
list_models = [
    'lda', 'lightgbm','lr', 
    'gbc', 'rf', 'dt', 'nb']

best = s.compare_models(include=list_models, sort='f1')
```

```python
# creating and tuning best model
gbc = s.create_model('gbc', fold=5)
tuned_model = s.tune_model(gbc, optimize='f1', fold=5)
```

```python
# printing default model params
print(gbc, '\n')

# printing tuned model params
print(tuned_model)
```

### Analyzing Results

```python
# plotting AUC
s.plot_model(tuned_model, plot='auc')
```

```python
# plotting Confusion Matrix
s.plot_model(tuned_model, plot='confusion_matrix')
```

```python
# plotting learning curve
s.plot_model(tuned_model, plot='learning')
```

```python
# plotting top features
s.plot_model(tuned_model, plot='feature')
```
